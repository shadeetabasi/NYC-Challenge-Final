```python
# Import dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
#loading restaurants CSV
real_estate_path ="re_data_clean.csv"
real_estate_df = pd.read_csv(real_estate_path)
real_estate_df.head()
```

```python
clean_real_estate_df = real_estate_df.drop(['index'], axis=1)
clean_real_estate_df
```

```python
#Reset index name to load into SQL database
clean_real_estate_df.index.name = "index"
clean_real_estate_df
```

```python
#connect engine
###note to group, update local host location and name of database to reflect what is on your system
connection_string = "postgres:postgres@localhost:5432/final_project"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
#checking tables names 
engine.table_names()
```

```python
clean_real_estate_df.to_sql(name='real_estate', con=engine, if_exists='append', index=True)
```
